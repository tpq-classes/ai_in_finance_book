## The Market

```python
S = np.array((20, 5))
B = np.array((11, 11))
```

```python
S0 = 10
B0 = 10
```

```python
M = np.array((S, B)).T
M
```

```python
M0 = np.array((S0, B0))
M0
```

## Option Pricing By Replication

```python
K = 14.5
```

```python
C = np.maximum(S - K, 0)
C
```

```python
phi = np.linalg.solve(M, C)
phi
```

```python
C0 = np.dot(M0, phi)
C0
```

## Option Pricing by Simple Neural Network

```python
l0 = M
```

```python
w = np.random.random(2)
w
```

```python
l1 = np.dot(l0, w)
l1
```

```python
e = l1 - C
e
```

```python
alpha = 0.01  # learning rate
```

```python
u = alpha * e
u
```

```python
w -= u
w
```

```python
w = np.random.random(2)
for _ in range(200):
    l1 = np.dot(l0, w)  # output layer
    e = l1 - C  # error values
    u = alpha * e  # update values
    w -= u  # weights update
```

```python
l1.round(6)
```

```python
(e ** 2).mean()  # MSE
```

```python
w
```

```python
C0_ = np.dot(M0, w)
C0_
```

```python
C0_ - C0
```
